```python
import seaborn as sns     # Seaborn is a powerful data visualization library built on top of Matplotlib.
                          #It is widely preferred in the industry for its ease of use and superior capabilities.
```

```python
                              # Load the "tips" dataset as 't'     (task 1)
t = sns.load_dataset("tips")  #loads the data set
t
```

```python
                                                           # Check the categorical columns in the dataset   (task 2)
categorical_columns = ['sex', 'day', 'time', 'size']

for column in categorical_columns:
    print(f"Unique values in {column}: {t[column].unique()}")  # Print unique values in categorical columns: sex, day, time, and size.
```

```python
                                            # Create a strip plot for gender-wise tips  (task 3)
sns.stripplot(data= t, x ='sex', y ='tip') 
```

Insights:
    Tips are evenly distributed among genders, with some variations.
    A few customers from both genders leave significantly larger tips.

```python
                                               # Create a strip plot for gender-wise total bill amounts (task 4)
sns.stripplot(data=t ,x='sex', y='total_bill' )
```

Insights:
    Male customers generally have slightly higher total bills.
    Outliers exist in both groups with notably higher bills.

```python
                                            # Create a time vs. total bill strip plot with 'sex' as the hue   (task 5)
sns.stripplot(data=t, x='time', y='total_bill', hue='sex', dodge=True)  #Data points for each category in the 'hue' variable
                                           #(e.g., male and female) are placed side by side but visually separated on the plot.
```

Insights:
Dinners tend to have higher total bills compared to lunches.
The plot helps observe bill distribution across different times and genders.
